```python
dictA = data2.to_dict('list');
print(dictA);

# count the number of true/false in Returned data for overall

import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

data_types = ['RNA_seq','DNA_seq','Methyl_seq']

def dict_extract(datatype, dic):  ### count the number of true/false and return a dictionary for each data type 
    list_return = dic[datatype]
    dict_return = dict((x,list_return.count(x)) for x in [True,False])
    return dict_return.values()

def dict_extract_all(datatype, dic):  ### count the number of true/false and return a dictionary for all data types 
    list_return_all = []
    for i in datatype:
        i = join(i + '_Returned')
        list_return_all += dic[i]
    dict_return_all = dict((x,list_return_all.count(x)) for x in [True,False])
    return dict_return_all.values()

#plot
langs = ['Returned','Failure Returned']
#colors = ['blue', 'red']

fig1 = plt.figure(figsize=(20,10), constrained_layout=True)
gs0 = GridSpec(1, 2, figure=fig1)


# plot overall
ax = fig1.add_subplot(gs[:, 0])
ax.pie(dict_extract_all(data_types, dictA), labels= langs, autopct='%.1f%%',radius = 1.2)
ax.set_title('OVERALL RETURNED RATE')

# plot overall by data type
gs01 = gs0[1].subgridspec(1, len(data_types))
n = 0
for i in data_types:
    name = join(i + "_Returned")
    ax = join('ax' + str(n))
    ax = fig1.add_subplot(gs01[0, n])
    ax.pie(dict_extract(name,dictA), labels=langs, autopct='%.1f%%')
    ax.set_xlabel(i)
    n += 1
ax.set_title('OVERALL RETURNED RATE BY DATA TYPE')    

plt.show()
```

```python
import yaml

from bokeh.layouts import column
from bokeh.models import ColumnDataSource, Slider
from bokeh.plotting import figure
from bokeh.themes import Theme
from bokeh.io import show, output_notebook

from bokeh.sampledata.sea_surface_temperature import sea_surface_temperature

output_notebook()
```

```python
def bkapp(doc):
    df = sea_surface_temperature.copy()
    source = ColumnDataSource(data=df)

    plot = figure(x_axis_type='datetime', y_range=(0, 25),
                  y_axis_label='Temperature (Celsius)',
                  title="Sea Surface Temperature at 43.18, -70.43")
    plot.line('time', 'temperature', source=source)

    def callback(attr, old, new):
        if new == 0:
            data = df
        else:
            data = df.rolling('{0}D'.format(new)).mean()
        source.data = ColumnDataSource.from_df(data)

    slider = Slider(start=0, end=30, value=0, step=1, title="Smoothing by N Days")
    slider.on_change('value', callback)

    doc.add_root(column(slider, plot))

    doc.theme = Theme(json=yaml.load("""
        attrs:
            Figure:
                background_fill_color: "#DDDDDD"
                outline_line_color: white
                toolbar_location: above
                height: 500
                width: 800
            Grid:
                grid_line_dash: [6, 4]
                grid_line_color: white
    """, Loader=yaml.FullLoader))
```
